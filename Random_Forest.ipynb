```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://raw.githubusercontent.com/FazlyRabbiBD/Data-Science-Book/master/data-drugs.csv'
df = pd.read_csv(url)
df.head()
```

```python
# setting dummy vaalues to the categorical variable  (Sex, BP, Cholesterol) using map function

d = {'M':1, 'F':0}
df['Sex'] = df['Sex'].map(d)

d = {'NORMAL':1, 'LOW':0, 'HIGH':2}
df['BP'] = df['BP'].map(d)

d = {'NORMAL':0, 'HIGH':1}
df['Cholesterol'] = df['Cholesterol'].map(d)
```

```python
df.sample(10)
```

```python
# feature and target variabe define
features = ['Age', 'Sex', 'BP', 'Cholesterol','Na_to_K']
x = df[features]
y = df['Drug']
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

# we will use 10 trees for for a random forest
rf = RandomForestClassifier(n_estimators=10, max_depth=5, random_state=0)
rf.fit(x_train, y_train)
```

```python
# prediction and accuracy check

predictions = rf.predict(x_test)
print('Classification Report \n:', classification_report(y_test, predictions))
print('\n Accuracy Score: ', accuracy_score(y_test, predictions))
print('\n Confusion Matrix \n:', confusion_matrix(y_test, predictions))
```
